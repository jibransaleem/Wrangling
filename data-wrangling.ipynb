```python
# Make a seprate column for  Role Entry 
import re
pattern = r'^(Senior|Sr\.?|Junior|Jr\.?)\s*'
exp  = data[data['Job Title'].str.contains(pattern)]
data['Role Entry'] = data['Job Title'].str.extract(pattern)


```

```python
#Job title has Exp level hidden data , we have extract that data  , now remove that information from this column
data['Job Title'] = data['Job Title'].str.replace(pattern, '', regex=True)
```

```python
#Data Descriptipn has \n in it , remove them
data['Job Description'] = data['Job Description'].apply(lambda x: x.replace('\n',''))
```

```python
temp = data['Salary Estimate'].str.split('(')
salary = temp.apply(lambda x : x[0])
source = temp.apply(lambda x:x[1])
data['Source'] = source.str.split(' ').apply(lambda x : x[0])
data['Salary Estimate'] = salary
```

```python
#Make a seprate column f
def Setter(x):
    if len(x)==1:
        return None
    else:
        return x[1]
        
data['Rating OF Company'] = data['Company Name'].str.split('\n').apply(lambda x : Setter(x))
```

```python
data.drop(columns =['Rating OF Company'])
```

```python
#Changing the Company Name column with company name only 
data['Company Name']= data['Company Name'].str.split('\n').apply(lambda x: x[0])
```

```python
data.drop(data.columns[-1], axis=1, inplace=True)
data.drop(data.columns[-1], axis=1, inplace=True)
```

```python
data['Revenue'] = data['Revenue'].replace('Unknown / Non-Applicable',np.nan)
```
